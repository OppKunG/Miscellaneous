```python
import pandas as pd
from selenium import webdriver

# Source: Krungsri

def get_currency():
    driver = webdriver.Chrome()
    driver.get('https://www.krungsri.com/th/exchangerate/Todayrates')
    return driver


driver = get_currency()
```

```python
def get_currency_df(driver):
    data = driver.page_source
    data_df = pd.read_html(data)[1]
    del data_df['SELL',      'SELL']

    clean_column = []
    for c in data_df.columns:
        clean_column.append(c[1])
    data_df.columns = clean_column

    delete_columns = ['COUNTRY', 'DENOM.', 'SIGHT', 'T/T']
    for d in delete_columns:
        del data_df[d]

    data_df.columns = data_df.columns.str.replace('COUNTRY.1', 'COUNTRY')
    data_df = data_df.set_index('COUNTRY')
    return data_df


data_df = get_currency_df(driver)
data_df
```
