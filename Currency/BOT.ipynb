```python
import pandas as pd
from selenium import webdriver

# Source: Bank Of Thailand

def get_currency():
    driver = webdriver.Chrome()
    driver.get('https://www.bot.or.th/thai/_layouts/application/exchangerate/exchangerate.aspx')
    return driver

driver = get_currency()
```

```python
from datetime import datetime


def get_currency_df(driver):
    data = driver.page_source

# 1. อัตราแลกเปลี่ยนถัวเฉลี่ยที่ธนาคารพาณิชย์ใช้ซื้อขายกับลูกค้า
# 2. อัตราในตลาดต่างประเทศ (ทอมสันรอยเตอร์) คำนวณผ่านอัตราซื้อขายเงินดอลลาร์ สรอ. ในตลาดกรุงเทพฯ

    country = []
    for i in range(0, 19):
        country.append(pd.read_html(data)[6][1][12+i])  # 1
    for i in range(0, 29):
        country.append(pd.read_html(data)[7][1][34+i])  # 2

    currency = []
    for i in range(0, 19):
        currency.append(pd.read_html(data)[6][2][12+i])  # 1
    for i in range(0, 29):
        currency.append(pd.read_html(data)[7][2][34+i])  # 2

    exchange_rate = []
    for i in range(0, 19):
        exchange_rate.append(pd.read_html(data)[6][5][12+i])  # 1
    for i in range(0, 29):
        exchange_rate.append(pd.read_html(data)[7][4][34+i])  # 2

    return country, currency, exchange_rate


def generate_dict(select_data):
    result_dict = {}
    for i in range(0, len(country)):
        data_df = select_data[i]
        result_dict[country[i]] = data_df

    return result_dict


def generate_df(result_dict, select_row):
    result_df = pd.DataFrame([result_dict])
    result_df = result_df.transpose()
    result_df.columns = [select_row]

    return result_df


timestamp = datetime.now()

country, currency, exchange_rate = get_currency_df(driver)


currency_df = generate_df(generate_dict(currency), 'สกุลเงิน')
exchange_df = generate_df(generate_dict(exchange_rate), 'อัตราแลกเปลี่ยน')

result_df = currency_df
result_df['อัตราแลกเปลี่ยน'] = exchange_df

print('Date:', timestamp)
result_df
```
